### Hashtags

The idea here is that when an event is happening and people are having a conversation about it on Twitter, a set of uniform hashtags that represent the event spontaneously evolves. Twitter users then use those hashtags to communicate with one another. Some hashtags, like `#RT` for "retweet" or just `#retweet` are used frequently and don't tell us much about what is going on. But a sudden appearance of a hashtag like `#oscars` probably indicates that the oscars are underway. ### User-defined functions

A neat trick of spark dataframes is that you can essentially use something very much like an RDD `map` method but without switching to the RDD. If you are familiar with database languages, this works very much like e.g. a user-defined function in SQL. 

So, for example, if we wanted to make a user-defined python function that returns the hashtags in lowercase, we could do something like this:

````python
@functions.udf
def lowercase(text):
    """Convert text to lowercase"""
    return text.lower()
````

The @functions.udf is a "decorator" -- this is really handy python syntactic sugar and in this case is equivalent to:

```
def lowercase(text):
    return text.lower()
    
lowercase = functions.udf(lowercase)
```

It basically takes our function and adds to its functionality. In this case, it registers our function as a pyspark dataframe user-defined function (UDF).

Using these UDFs is very straightforward and analogous to other Spark dataframe operations. For example:

````python
df.select(lowercase(df.hashtag)).show()
````

Using a framework like Spark is all about understanding the ins and outs of how it functions and knowing what it offers. One of the cool things about the dataframe API is that many functions are already defined for you (turning strings into lowercase being one of them).

### Using `Window` functions 

Window functions are another awesome feature of dataframes. They allow users to accomplish complex tasks using very concise and simple code. 

Above we computed just the hashtag that had the most occurrences on *any* day. Now lets say we want to know the top tweets for *each* day.  

This is a non-trivial thing to compute and requires "windowing" our data. I recommend reading this [window functions article](https://databricks.com/blog/2015/07/15/introducing-window-functions-in-spark-sql.html) to get acquainted with the idea. You can think of a window function as a fine-grained and more flexible `groupBy`. 

There are two things we need to define to use window functions:

1. the "window" to use, based on which columns (partitioning) and how the rows should be ordered 
2. the computation to carry out for each windowed group, e.g. a max, an average etc.

Lets see how this works by example. We will define a window function, `daily_window` that will partition data based on the `day` column. Within each window, the rows will be ordered by the daily hashtag count that we computed above. Finally, we will use the rank function **over** this window to give us the ranking of top tweets. 

In the end, this is a fairly complicated operation achieved in just a few lines of code! (can you think of how to do this with an RDD??)

````python
from pyspark.sql import Window
````

First, we specify the window function and the ordering:

````python
daily_window = Window.partitionBy('day').orderBy(functions.desc('count'))
````

The above window function says that we should window the data on the `day` column and order it by count. 

Now we need to define what we want to compute on the windowed data. We will start by just calculating the daily ranking of hashtags, so we can use the helpful built-in `functions.rank()` and sort:

````python
daily_rank = functions.rank() \
                      .over(daily_window) \
                      .alias('rank')      
````

````python
weekly_w = Window.partitionBy('week').orderBy(functions.desc('week'))
````

We will opt for the chi-squared measure of trendiness between different hashtags, with a meaning higher chi-squared higher trendiness of the hashtag is compared to the rest of hashtags. When we are speaking trendiness, we speak about weekly-trendiness hence we will compare successive weeks hashtags. We treat previous week hashtag as expected or prior distribution, and the current week hashtag's mentions as observed distirbution. For more details consult slide 9-12 from Source:http://blogs.ischool.berkeley.edu/i290-abdt-s12/files/2012/08/Kostas_Trends_Sept_13_2012.pdf



Here we will illustrate through pictures steps will follow:


, 
Measure of trendiness for hashtag1 and total amount of the smoothed mentions per week in this examplewould be:
    

````python
# here we let renormalize the mentions such that total number of them, is constant for all weeks, and let it be N
# that is important to be able to calculate chi-square as measure of trendiness

# percent_mentions, are simple percent of the mentions for one hashtag of mentions of all hashtags for that week
# smoothed_num_mentions has been just added to each of these percent, just one percent, 
# so we avoid division by 0, = situation when in previous week there were no mentions of the hashtag 
# for which we want to measur trendiness in this week 
sum_by_week = sum_by_week.withColumn('window_sum', functions.sum("mentions").over(weekly_w))\
    .withColumn('percent_mentions', functions.col('mentions') / functions.col('window_sum')*100)\
    .withColumn('smoothed_num_mentions', functions.col('percent_mentions')+1)\
    .drop('window_sum')
sum_by_week.show(5)
````

````python
# from the slides, this is N, or the total amount of smoothed_num_mentions for any week
# it is equal to 100 (before adding 1 as smoothing we have 100 amount of hash in each week, they were percent), and 0 for each hash
# to take care of not having any hashtag in week before the one for which we calculate trendiness by formula (E-O)**2/E + (E-O)**2/(N-E)
total_number_of_hash = sum_by_week.select("hashtag").distinct().count() + 100
total_number_of_hash
````

````python
# make a dataframe, where rows are hashtags, columns are weeks and values respresent smoothed_num_mentions of hashtag for the week
hash_by_week = sum_by_week.groupby('hashtag').pivot('week').max('smoothed_num_mentions').na.fill(1)
````

````python
def calculate_trendiness(df, col_prev_week, col_curr_week, smoothed_number_of_hash):
    """
    df: pivoted dataframe with col_prev_week with number of mentions that corresponds to expected distribution of tweets
    col_prev_week: column which contains hashtags smoothed mentions from previous week, or our expected distribution, E
    col_curr_week: column which contains hashtags smoohted mentions from current week, or our observed distribution, O
    smoothed_number_of_hash: total number of smoothed mentions per each week, N
    
    based on formula for chi-square to determine trendiness, see slide 12 from link 
    (E-O)**2/E + (E-O)**2/(N-E) = (E-O)**2*N/E*(N-E)
    
    """
    denom_term = functions.pow(df[col_curr_week]-df[col_prev_week],2)*smoothed_number_of_hash
    # only if df[coll_cur_week]>df[col_prev_week] we observe trendiness, otherwise trendiness is 0
    ascending_hashtag = functions.when(df[col_curr_week]>df[col_prev_week], 1).otherwise(0)
    denom = denom_term * ascending_hashtag
    nom = df[col_prev_week]*smoothed_number_of_hash - functions.pow(df[col_prev_week],2)
    return denom / nom
````

Calculating trendiness of hashtags for each week, requires calling function on successive weeks, here we find their column names

````python
# all expected weeks:
columns_prev_weeks = list(hash_by_week.columns)[1:-1]
# all observed weeks:
columns_curr_weeks = list(hash_by_week.columns)[2:]
````

````python
# applying previous function to calculate amount of trendiness for each tweet each week, and store it in new df
sc = spark.sparkContext
chi_square_by_weeks = spark.createDataFrame(sc.emptyRDD(), hash_by_week.schema)
for col_prev_week, col_curr_week in zip(columns_prev_weeks, columns_curr_weeks):
        chi_square_by_weeks = hash_by_week.withColumn(col_curr_week, \
                              calculate_trendiness(hash_by_week, col_prev_week, col_curr_week, total_number_of_hash))
````

Here we will try to unpivot the dataframe, with hashtags, week and trendiness measure as only columns, using melt function taken from Source: https://stackoverflow.com/questions/41670103/how-to-melt-spark-dataframe

````python
from pyspark.sql.functions import array, col, explode, lit, struct
from pyspark.sql import DataFrame
from typing import Iterable 

def melt(
        df: DataFrame, 
        id_vars: Iterable[str], value_vars: Iterable[str], 
        var_name: str="variable", value_name: str="value") -> DataFrame:
    """Convert :class:`DataFrame` from wide to long format."""

    # Create array<struct<variable: str, value: ...>>
    _vars_and_vals = functions.array(*(
        functions.struct(functions.lit(c).alias(var_name), functions.col(c).alias(value_name)) 
        for c in value_vars))

    # Add to the DataFrame and explode
    _tmp = df.withColumn("_vars_and_vals", functions.explode(_vars_and_vals))

    cols = id_vars + [functions.col("_vars_and_vals")[x].alias(x) for x in [var_name, value_name]]
    return _tmp.select(*cols)
````

````python
# we take from second week, since we do not have trendiness for first week
trendiness_per_week = melt(chi_square_by_weeks, id_vars = ['hashtag'], value_vars = chi_square_by_weeks.columns[2:])
# renaming column names
trendiness_per_week = trendiness_per_week.select("hashtag", functions.col("variable").alias("week"), functions.col("value").alias("trendiness"))
trendiness_per_week.show()
````

````python
from pyspark.ml.clustering import KMeans
from pyspark.ml.evaluation import ClusteringEvaluator
from pyspark.sql.types import BooleanType
from pyspark.sql.functions import udf
from pyspark.ml.linalg import Vectors, VectorUDT
import numpy as np
import math
````

````python
#filter out hashtags with less than 1000 uses to reduce noise
def ith_(v):
    agg= 0.0
    for x in v:
        try:
            agg += float(x)
        except ValueError:
            agg = agg
    return agg

ts_features_filt = ts_features_df.select("hashtag", "features")\
.filter(udf(lambda x: ith_(x) > 1000, BooleanType())("features"))


# rescale the hashtags according to maximum
def max_rescale(vector):
    max_val = max(vector)                              # maximum value in the vector
    return Vectors.dense([x/max_val for x in vector])    # scale it

# create a udf for the scaler
max_sca_udf = udf(max_rescale, VectorUDT())
ts_features_filt_scal = ts_features_filt.withColumn("features", max_sca_udf(ts_features_filt["features"]))
````

The created Dataframe first is filtered to contain only the hashtags with more than 1000 mentions over the total days. This will lead to reduction of noise in the clustering due to hashtags that have rarely been used. A hashtag occuring only once every now and then will thus not interfere with the clustering and only real trends are modeled.

Furthermore we rescale the hashtags according to the counts on the most cited day to better capture temporal trends in the hashtags and reduce the influence of overall counts of a hashtag.

The filter and processing we apply leads us to losing some features of the data, nevertheless we argue that they are necessairy, as we are mostly interested in forming groups of temporally aligned hashtags.

````python
def silu_gen(k):
    # Train a k-means model.
    kmeans = KMeans().setK(k).setSeed(1)
    model = kmeans.fit(ts_features_filt_scal)
    predictions = model.transform(ts_features_filt_scal)

    evaluator = ClusteringEvaluator()
    silhouette = evaluator.evaluate(predictions)
    return silhouette

x = np.logspace(2,4,20)
x = np.array([math.floor(xs) for xs in x])

y = np.array([silu_gen(xs) for xs in x])
plt.plot(x,y)
````

We use the well known elbow method to find a appropriate k-value for the clustering. The applied evaluation function in pyspark gives the silhouette score of the clustering. Where score value from [-1, 1] can be achieved. We see that for the 20 different values of k tried, the curve is generally increasing with k. A plateau value is found at about 6000, thus, we use a k = 6000 for the clustering as it is the minimal k to still achieve high scores.

````python
# Train a k-means model.
kmeans = KMeans().setK(6000).setSeed(1)
model = kmeans.fit(ts_features_filt_scal)
predictions = model.transform(ts_features_df)
````
